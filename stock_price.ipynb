```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset
df = pd.read_csv("stock_prices.csv")
```

```python
# Convert date column to datetime object
df['date'] = pd.to_datetime(df['date'])
```

```python
# Set date as the index
df.set_index('date', inplace=True)
```

```python
# Calculate the rolling 30-day moving average
rolling_avg = df['closing_price'].rolling(window=30).mean()
```

```python
# Plot the closing price and rolling average
plt.plot(df.index, df['closing_price'], label='Closing Price')
plt.plot(df.index, rolling_avg, label='30-day Moving Average')
plt.xlabel('Date')
plt.ylabel('Closing Price')
plt.title('Stock Prices')
plt.legend()
plt.show()
```
