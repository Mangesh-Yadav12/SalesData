```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the dataset
data = pd.read_csv('sales_data.csv')
```

```python
# Calculate total revenue for each product
product_revenue = data.groupby('product_id')['revenue'].sum().sort_values(ascending=False)
```

```python
# Select top 10 products by revenue
top_10_products = product_revenue.head(10)
```

```python
# Create a bar plot
plt.bar(top_10_products.index, top_10_products.values)
plt.xlabel('Product ID')
plt.ylabel('Total Revenue')
plt.title('Top 10 Products by Revenue')
plt.xticks(rotation=45)
plt.show()
```
